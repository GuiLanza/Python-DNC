```python
df_filtrado = df_sem_nulos.query('Quantity > 0')

# Exibir o dataset filtrado
print(df_filtrado)
```

```python
linhas_duplicadas = df_filtrado.duplicated()

# Verificar se existem linhas duplicadas
if linhas_duplicadas.any():
    print("Existem linhas duplicadas no DataFrame.")
else:
    print("Não existem linhas duplicadas no DataFrame.")
```

```python
# Remover as linhas duplicadas
df_sem_duplicatas = df_filtrado.drop_duplicates()

# Exibir o DataFrame sem linhas duplicadas
print(df_sem_duplicatas)
```

```python
df_sem_duplicatas['CustomerID'] = df_sem_duplicatas['CustomerID'].astype(int)
```

```python
df_sem_duplicatas.dtypes
```

```python
outliers = df_sem_duplicatas[(df_sem_duplicatas['Quantity'] > 10000) | (df_sem_duplicatas['UnitPrice'] > 5000)]
print("Outliers:")
print(outliers)

# Remover os outliers extremos
df_sem_outliers = df_sem_duplicatas[(df_sem_duplicatas['Quantity'] <= 10000) & (df_sem_duplicatas['UnitPrice'] <= 5000)]
print("\nDataFrame sem outliers extremos:")
print(df_sem_outliers)
```

```python
# Criar a coluna 'Preço Total' com base nas colunas 'Quantidade' e 'Preço Unitário'
df_sem_outliers['Total price'] = df_sem_outliers['Quantity'] * df_sem_outliers['UnitPrice']

# Exibir o DataFrame com a coluna 'Preço Total' adicionada
print(df_sem_outliers)
```

```python
# Calcular a data da última compra
data_ultima_compra = df_sem_outliers['data'].max()

# Exibir a data da última compra
print("Data da última compra:", data_ultima_compra)
```

```python
# Agrupar os dados por país e somar o valor de vendas
vendas_por_pais = df_sem_outliers.groupby('Country')['Total price'].sum()

# Selecionar os top 10 países com maior valor em vendas
top_10_paises = vendas_por_pais.nlargest(10)

# Plotar o gráfico de barras
plt.bar(top_10_paises.index, top_10_paises.values)
plt.xlabel('Country')
plt.ylabel('Total price')
plt.title('Top 10 Países com Maior Valor de Venda')
plt.xticks(rotation=45)
plt.show()
```

```python
vendas_por_produto = df_sem_outliers.groupby('Description')['Quantity'].sum()

# Selecionar os top 10 produtos mais vendidos
top_10_produtos = vendas_por_produto.nlargest(10)

# Plotar o gráfico de barras
plt.bar(top_10_produtos.index, top_10_produtos.values)
plt.xlabel('Description')
plt.ylabel('Quantity')
plt.title('Top 10 Produtos Mais Vendidos')
plt.xticks(rotation=90)
plt.show()
```

```python
vendas_por_mes = df_sem_outliers.groupby(df_sem_outliers['data'].dt.month)['Total price'].sum()

# Plotar o gráfico de linha
plt.plot(vendas_por_mes.index, vendas_por_mes.values)
plt.xlabel('Mês')
plt.ylabel('Valor de Venda Total')
plt.title('Valor de Venda Total por Mês')
plt.xticks(range(1, 13))
plt.show()
```

```python
top_10_paises = vendas_por_pais.nlargest(10).index

# Filtrar o DataFrame apenas para os top 10 países
df_top_10_paises = df_sem_outliers[df_sem_outliers['Country'].isin(top_10_paises)]

# Agrupar os dados por mês e país, somando o valor de vendas
vendas_por_mes_pais = df_top_10_paises.groupby([df_top_10_paises['data'].dt.month, 'Country'])['Total price'].sum()

# Reformatar os dados em um DataFrame
df_vendas_por_mes_pais = vendas_por_mes_pais.unstack(level='Country')

# Plotar o gráfico de linhas
df_vendas_por_mes_pais.plot(kind='line', marker='o')
plt.xlabel('Mês')
plt.ylabel('Valor de Venda Total')
plt.title('Valor de Venda Total por Mês e por País (Top 10)')
plt.xticks(range(1, 13))
plt.legend()
plt.show()
```

```python
# Agrupar os dados por cliente e pedido/compra (InvoiceNo)
df_cliente_pedido = df_sem_outliers.groupby(['CustomerID', 'InvoiceNo']).agg({'data': 'max', 'Total price': 'sum'}).reset_index()

# Exibir o DataFrame com a data e o preço total do pedido/compra para cada cliente
print(df_cliente_pedido)
```

```python
ultima_compra = df_sem_outliers['data'].max()
df_cliente_pedido['Recência'] = (ultima_compra - df_cliente_pedido['data']).dt.days

# Calcular a frequência (F) e o ticket médio (M)
df_rfm = df_cliente_pedido.groupby('CustomerID').agg({'Recência': 'min', 'InvoiceNo': 'count', 'Total price': 'mean'}).reset_index()
df_rfm.columns = ['CustomerID', 'Recência', 'Frequência', 'Ticket Médio']

# Exibir o DataFrame com o RFM calculado para cada cliente
print(df_rfm)
```
